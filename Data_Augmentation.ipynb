```python
from keras.preprocessing import image #behaves like opencv in keras
from keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('/content/train/dog.jpeg',target_size=(200,200))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
type(img)
```

```python
datagen = ImageDataGenerator(
    rotation_range = 30, #angle of rotation of image
    shear_range = 0.2, #distortion
    zoom_range=0.2,#zoom in zoom out randomly
    horizontal_flip= True,
    width_shift_range=0.2, #shifting image by width
    height_shift_range = 0.2,
)
```

```python
img = image.img_to_array(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1,200,200,3) #dividing image into batches while transformation
```

```python
type(img) #convert PIL into numpy
```

```python
i = 0
#use flow when we use single photo in numpy format
for ouput in datagen.flow(input_batch, batch_size=1,save_to_dir='aug'):
  i = i + 1;

  #stopping when 10 images are formed
  if i == 10:
    break
```
